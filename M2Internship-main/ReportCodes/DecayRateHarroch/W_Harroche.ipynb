```julia
using PyPlot
```

```julia
J = 10
#M = range(-J, J, 100)
M = [-J:1:J;];
```

```julia
close("all")
figure()
PyPlot.rc("font", size = 12)
scatter(M, (J .+ M).*(J .- M .+1), label="Collective decay")
scatter(M, M .+J, label="SE")
plt.gca().invert_xaxis()
PyPlot.xticks(M[1:2:end])
axvline(0, color="gray")
pygui(false)
xlabel(L"M", fontsize=16)
ylabel(L"\mathcal{W}^{\ J, M}", fontsize=16)
legend(loc="center")
savefig("/Users/remy/Documents/Remy/ENSL_2024_2025/Stage_M2/Rapport_Stage_M2/Figures/DecayRateJM.pdf", bbox_inches="tight")
#pygui(true); show()
```
